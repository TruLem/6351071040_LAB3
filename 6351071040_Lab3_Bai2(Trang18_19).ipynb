# 6. Chọn thuộc tính Tuổi và Số điểm mua sắm để gom cụm. Sử dụng phương pháp khuỷu tay (ELBOW method) để xác định số cụm cần thiết. Tiến hành gom cụm bằng thuật toán k-Means với số cụm vừa xác định.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

print("Huỳnh Thị Trúc Lam - 6351071040")

# Select the features for clustering
X = df[['Age', 'Spending Score (1-100)']]

# Calculate WSS for a range of k values
wss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10) # n_init for modern sklearn versions
    kmeans.fit(X)
    wss.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wss, marker='o')
plt.title('Phương pháp khuỷu tay để xác định số cụm')
plt.xlabel('Số lượng cụm (k)')
plt.ylabel('Tổng bình phương khoảng cách trong cụm (WSS/Inertia)')
plt.grid(True)
plt.show()
```

# 7. Biểu diễn kết quả gom cụm trên bằng biểu đồ.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

print("Huỳnh Thị Trúc Lam - 6351071040")

# Optimal number of clusters identified from Elbow Method
k = 4

# Initialize and fit K-Means model
kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X)

# Add cluster labels to the DataFrame for easier plotting
df['Cluster_Age_Spending'] = y_kmeans

# Visualize the clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Age', y='Spending Score (1-100)', hue='Cluster_Age_Spending', data=df,
                palette='viridis', s=100, alpha=0.8, legend='full')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X', label='Tâm cụm')
plt.title(f'Gom cụm khách hàng (Tuổi vs Số điểm mua sắm) với {k} cụm')
plt.xlabel('Tuổi')
plt.ylabel('Số điểm mua sắm (1-100)')
plt.legend()
plt.grid(True)
plt.show()
```

# 8. Đánh giá kết quả gom cụm bằng Hệ số Dáng điệu (Silhouette coefficient).

```python
from sklearn.metrics import silhouette_score

print("Huỳnh Thị Trúc Lam - 6351071040")

# Calculate the Silhouette Score
silhouette_avg = silhouette_score(X, y_kmeans)

print(f"\nHệ số Dáng điệu trung bình cho các cụm là: {silhouette_avg:.3f}")
```

## Gom cụm theo Thu nhập hằng năm và Số điểm mua sắm

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

print("Huỳnh Thị Trúc Lam - 6351071040")

# Select the features for clustering: Annual Income and Spending Score
X_income_spending = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Calculate WSS for a range of k values using the Elbow Method
wss_income_spending = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(X_income_spending)
    wss_income_spending.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wss_income_spending, marker='o')
plt.title('Phương pháp khuỷu tay cho Thu nhập và Điểm mua sắm')
plt.xlabel('Số lượng cụm (k)')
plt.ylabel('Tổng bình phương khoảng cách trong cụm (WSS/Inertia)')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import silhouette_score

print("Huỳnh Thị Trúc Lam - 6351071040")

# Optimal number of clusters identified from Elbow Method (k=5)
k_income_spending = 5

# Perform K-Means clustering
kmeans_income_spending = KMeans(n_clusters=k_income_spending, init='k-means++', random_state=42, n_init=10)
y_kmeans_income_spending = kmeans_income_spending.fit_predict(X_income_spending)

# Add cluster labels to the DataFrame for easier plotting
df['Cluster_Income_Spending'] = y_kmeans_income_spending

# Visualize the clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster_Income_Spending', data=df,
                palette='viridis', s=100, alpha=0.8, legend='full')
plt.scatter(kmeans_income_spending.cluster_centers_[:, 0], kmeans_income_spending.cluster_centers_[:, 1], s=300, c='red', marker='X', label='Tâm cụm')
plt.title(f'Gom cụm khách hàng (Thu nhập vs Số điểm mua sắm) với {k_income_spending} cụm')
plt.xlabel('Thu nhập hằng năm (k$)')
plt.ylabel('Số điểm mua sắm (1-100)')
plt.legend()
plt.grid(True)
plt.show()

# Calculate the Silhouette Score
silhouette_avg_income_spending = silhouette_score(X_income_spending, y_kmeans_income_spending)
print(f"\nHệ số Dáng điệu trung bình cho các cụm (Thu nhập vs Số điểm mua sắm) là: {silhouette_avg_income_spending:.3f}")
```

## Gom cụm theo Tuổi, Thu nhập hằng năm và Số điểm mua sắm

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

print("Huỳnh Thị Trúc Lam - 6351071040")

# Select the features for 3D clustering: Age, Annual Income, and Spending Score
X_3d = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Calculate WSS for a range of k values using the Elbow Method
wss_3d = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(X_3d)
    wss_3d.append(kmeans.inertia_)

# Plot the Elbow Method graph for 3D clustering
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wss_3d, marker='o')
plt.title('Phương pháp khuỷu tay cho Tuổi, Thu nhập và Điểm mua sắm')
plt.xlabel('Số lượng cụm (k)')
plt.ylabel('Tổng bình phương khoảng cách trong cụm (WSS/Inertia)')
plt.grid(True)
plt.show()
```

**Reasoning**:
Based on the Elbow plot, the optimal number of clusters for the 3D data ('Age', 'Annual Income (k$)', and 'Spending Score (1-100)') appears to be 6. I will now perform K-Means clustering with k=6, visualize the results in a 3D scatter plot, and calculate the Silhouette Coefficient.



```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import silhouette_score
from mpl_toolkits.mplot3d import Axes3D

print("Huỳnh Thị Trúc Lam - 6351071040")

# Optimal number of clusters identified from Elbow Method (k=6)
k_3d = 6

# Perform K-Means clustering
kmeans_3d = KMeans(n_clusters=k_3d, init='k-means++', random_state=42, n_init=10)
y_kmeans_3d = kmeans_3d.fit_predict(X_3d)

# Add cluster labels to the DataFrame for easier plotting
df['Cluster_Age_Income_Spending'] = y_kmeans_3d

# Visualize the clusters in 3D
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for each cluster
scatter = ax.scatter(df['Age'], df['Annual Income (k$)'], df['Spending Score (1-100)'],
                     c=df['Cluster_Age_Income_Spending'], cmap='viridis', s=100, alpha=0.7)

# Plot cluster centers
ax.scatter(kmeans_3d.cluster_centers_[:, 0], kmeans_3d.cluster_centers_[:, 1], kmeans_3d.cluster_centers_[:, 2],
           s=300, c='red', marker='X', label='Tâm cụm')

ax.set_title(f'Gom cụm khách hàng (Tuổi, Thu nhập, Điểm mua sắm) với {k_3d} cụm')
ax.set_xlabel('Tuổi')
ax.set_ylabel('Thu nhập hằng năm (k$)')
ax.set_zlabel('Số điểm mua sắm (1-100)')
ax.legend()
plt.show()

# Calculate the Silhouette Score
silhouette_avg_3d = silhouette_score(X_3d, y_kmeans_3d)
print(f"\nHệ số Dáng điệu trung bình cho các cụm (Tuổi, Thu nhập, Số điểm mua sắm) là: {silhouette_avg_3d:.3f}")
```
